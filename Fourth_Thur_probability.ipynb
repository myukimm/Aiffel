### 상호 배타적(배반) 사건과 독립사건의 차이

상호 배타적 사건: 시행의 모든 시도에서 A 혹은 B사건이 발생했을 때, 다른 하나의 사건이 발생하지 않을 경우 A와 B는 서로 배타적 입니다.

독립 사건 : 두 사건 A, B에 대해서 한 사건이 다른 하나의 사건에 영향을 주지 않을 때, 'A와 B는 독립이다'라고 합니다.

#### 독립 사건인지 배반 사건인지 구분하는 꿀팁

P(A), P(B) 교집합이 = P(A) * P(B)이 성립되면 독립니다.

### 정리

X(w)X(w)는 '확률 변수'라는 함수이며, ww라는 표본 공간의 한 표본점(시행을 통해 얻을 수 있는 경우 중 하나)을 실수 직선으로 맵핑해줍니다.

xx는 ww라는 표본점이 확률 변수 XX에 의해 대응된 하나의 실수 입니다.

ex) Ax는 사건이므로 확률을 가지고 있다. 즉, p = P(Ax) = P(X = x)

확률로 표현

P(X≤x)는 XX의 값이 어떤 실수 값 xx값보다 작거나 같을 확률입니다.

P(X > x)P(X>x)는 XX의 값이 어떤 실수 값 xx보다 클 확률입니다. 이는 1 - P(X \leq x)1−P(X≤x)와 같습니다.

P(a < X < b )P(a<X<b)는 XX의 값이 aa와 bb 사이에 있을 확률

동전 던지기

(앞, 앞), (앞, 뒤), (뒤,앞), (뒤, 뒤)

이것을 위에 나열한 순서대로 1, 2, 3, 4에 대응시키면 다음과 같이 표현할 수 있다.

X((앞,앞))=1,X((앞,뒤))=2,X((뒤,앞)=3,X((뒤,뒤))=4

image.png

{X = 3} = {x|X(w) = 3} = {(뒤, 앞)}

{X <= 3} = {x|X(w) <= 3} = {(앞, 앞), (앞, 뒤),(뒤, 앞)}

{X > 3} = {x|X(w) > 3} = {(뒤, 뒤)}

{1 < X < 4} = {x|1 < X(w) < 4} = {(앞, 뒤),(뒤, 앞)}

X의 값이 3일 확률은 P {X = 3} = 1/4

X의 값이 3 이하일 확률은 P {X<=3} = 3/4

X의 값이 3 초과일 확률은 P {X>3} = 1/4

X의 값이 1과 4사이일 확률은 P {1 < X <4} = 1/2

# 분포 함수

X는 확률 변수이고 x가 하나의 숫자라고 합시다. 우리는 사건을 {w|X(w) <= X} = {X <= x}로 표현할 수 있다. 이 때의 확률 P(X <= x)을 Fx(x)라 한다면 다음과 같이 표현할 수 있다.

Fx(x) = P(X<=x), -무한대 < x < 무한대

이것을 분포 함수(distribution function) 또는 누적 분포 함수(cumulative distribution function, CDF)라고 한다.

image.png

Fx(X)는 단조 증가함수(monotonic increasing function)이다. 즉, 어떤 수는 x1, x2에 대해 x1 <= x2이면 Fx(x1) <= Fx(x2)이다.

# 확률 및 확률분포

### 기댓값

이산 확률 변수 X의 기대값은 각 X의 값에 그 값에 해당하는 확률을 곱하여 모두 더한 값이 된다. 이 때 X의 값에 곱하는 확률은 흔히 X의 값에 대한 '가중치'를 곱한다고 표현

### 분산

값들의 평균을 구하고, 각 값에 평균을 빼준다. 편차 제곱의 평균을 구한다.

image.png

### 조건부 기댓값

가끔 우리는 '시험을 통과한' 학생들의 '평균' 시험 성적이나 '박사학위를 가진' 교수님들의 '평균' 나이를 알고 싶을 때가 있습니다. 이때 사용할 수 있는 것이 바로 조건부 기댓값(conditional expectation) 입니다.

단어의 의미를 좀 더 생각해보자면, '시험을 통과한' 그리고 '박사학위를 가진' 이라는 것이 '조건'이며, 이러한 조건하에 '평균'을 구하려고 하니 '기댓값'이라고 표현한 것입니다.

PX|A (x|A) = px(x) / P(A) 

주사위를 던진다고 했을 때, 짝수가 나온 경우의 결과값들의 평균

P(A) = 1/2

위 식에 따라 2px(X)

E[X|A] = 2 * 2 * 1/6 + 4 * 2 * 1/6 + 6 * 2 * 1/6 = 4

# 베르누이 시도와 베르누이 분포

베르누이 시도(Bernoullit trial)는 결과가 성공 혹은 실패 두가지로만 나오는 실험

P(성공) = p, 0 <= p <= 1

P(실패) = 1 - p

# 이항분포

독립적으로 n번의 베르누이 시도를 했고, 그 중 성공한 수를 X(n)로 표현할 때, 이 X(n)을 이항 확률 변수(binomial random variable)라고 한다. 이 때 이항 확률 변수 X(n)은 파라미터 (n, p)를 가지고 있다. 여기서 n은 시행한 횟수이고, p는 성공할 확률을 뜻한다.

```python
# 이항 분포 그림
import math
import matplotlib.pyplot as plt
from functools import reduce
import operator as op

n = 10 # 횟수
p = 0.5 # 성공할 확률

# 좌표 설정
plot_x_range = n
plot_y_range = 0.4

# 조합
def ncr(n, r):
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer // denom 

# 이항 분포 함수
def p_X(x,n,p):
    return (ncr(n, x))*(p**x)*((1-p)**(n-x)) 

# X값
X = [x for x in range(n+1)]

# 그래프 그리기
plt.figure(figsize=(10,10))
ax1 = plt.subplot(2, 1 ,1)
plt.xlim(0, plot_x_range)
plt.ylim(0,plot_y_range)
plt.title("Binomial distribution with several trials")

ax2 = plt.subplot(2, 1 ,2)
plt.xlim(0, plot_x_range)
plt.ylim(0,plot_y_range)
plt.title("Binomial distribution with several probabilities")

for different_n in [4+3*i for i in range(5)]:
    different_X = [x for x in range(different_n+1)]
    b_y = [p_X(x,different_n,p) for x in different_X]
    ax1.plot(different_X, b_y, label='n='+str(different_n))
    
for different_p in [(1+2*i)/10 for i in range(5)]:
    b_y = [p_X(x,n,different_p) for x in X]
    ax2.plot(X, b_y, label='p='+str(different_p))
    
ax1.legend(loc="best")
ax2.legend(loc="best")
plt.show()
```

# 정규 분포

연속 확률 변수 X의 분포

```python
# 정규 분포 그림
import math
import matplotlib.pyplot as plt

u = 0 # 평균
v = 1# 분산

# 좌표 설정
n = 1000
plot_x_range = 3
plot_y_range = 0.5

# 정규 분포 함수 (직관적 이해를 위해 수식을 그대로 표현)
def f_X(x,u,v):
    return 1/(math.sqrt(2*math.pi*v))*math.exp(-(x-u)**2/(2*v))

# X값
X = [-plot_x_range+2*plot_x_range*x/n for x in range(n+1)]

# 그래프 그리기
plt.figure(figsize=(10,10))
ax1 = plt.subplot(2, 1 ,1)
plt.xlim(-plot_x_range, plot_x_range)
plt.ylim(0,plot_y_range)
plt.title("Normal distribution with several means")

ax2 = plt.subplot(2, 1 ,2)
plt.xlim(-plot_x_range, plot_x_range)
plt.ylim(0,plot_y_range)
plt.title("Normal distribution with several variances")

for shifted_u in range(-2,3):
    n_y = [f_X(x, shifted_u, v) for x in X]
    ax1.plot(X, n_y, label='u='+str(shifted_u))
    
for different_v in [(8+10*i)/10 for i in range(5)]:
    n_y = [f_X(x, u, different_v) for x in X]
    ax2.plot(X, n_y, label='v='+str(different_v))
    
ax1.legend(loc="best")
ax2.legend(loc="best")
plt.show()
```

# 이항 분포와 정규 분포의 관계

```python
# 이항 분포와 정규 분포의 관계
import matplotlib.pyplot as plt
from functools import reduce
import operator as op
import math

# 조합
def ncr(n, r):
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer // denom 

# 이항 분포 함수
def p_X(x,n,p):
    return (ncr(n, x))*(p**x)*((1-p)**(n-x)) 

# 정규 분포 함수
def f_X(x,u,v):
    return 1/(math.sqrt(2*math.pi*v))*math.exp(-(x-u)**2/(2*v))


    
# 그래프 그리기
plt.figure(figsize=(10,20))

ax1 = plt.subplot(3, 1, 1)
plt.title("Binomial")
plt.xlim(0,25)
plt.ylim(0,0.35)

ax2 = plt.subplot(3, 1, 2)
plt.title("Normal")
plt.xlim(0,25)
plt.ylim(0,0.35)

X = [x for x in range(n)]

for i in range(1, 10):
    
    # 이항 분포 파라미터
    n=10*i
    p=1/8
    # 정규 분포 파라미터
    u=n*p
    v=n*p*(1-p)

    # 그래프
    b_y = [p_X(t, n, p) for t in X]
    n_y = [f_X(t, u, v) for t in X]

    ax1.plot(X, b_y, label='n='+str(n))
    ax2.plot(X, n_y, label='n='+str(n))

ax1.legend()
ax2.legend()
plt.show()
```
